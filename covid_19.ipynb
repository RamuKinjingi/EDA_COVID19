```python
# identifying the garbage value
for i in data.select_dtypes(include=['object']):
  print(data[i].value_counts())
```

```python
#outliers
numeric_cols = data.select_dtypes(include=np.number).columns
for col in numeric_cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = data[(data[col] < (Q1 - 1.5 * IQR)) | (data[col] > (Q3 + 1.5 * IQR))]
    print(f"{col}: {len(outliers)} outliers detected")
```

```python
# relationship analysis
data.select_dtypes(include=['number']).corr()
```

```python
# data visualization
```

```python
plt.figure(figsize=(14,6))
plt.bar(data['Name of State / UT'], data['Total Confirmed cases'], color='orange')
plt.title("Total Confirmed Cases by State (No Groupby)")
plt.xlabel("State / UT")
plt.ylabel("Total Confirmed Cases")
plt.xticks(rotation=90)
plt.show()
```

```python
latest = data.drop_duplicates(subset=['Name of State / UT'], keep='last')
```

```python
#Bar Chart
plt.figure(figsize=(12,5))
plt.bar(latest['Name of State / UT'], latest['Total Confirmed cases'], color='orange')
plt.title("Total Confirmed Cases by State - Bar Chart")
plt.xlabel("State / UT"); plt.ylabel("Total Confirmed Cases")
plt.xticks(rotation=90); plt.tight_layout(); plt.show()
```

```python
#Horizontal Bar Chart
plt.figure(figsize=(10,6))
plt.barh(latest['Name of State / UT'], latest['Total Confirmed cases'], color='teal')
plt.title("Total Confirmed Cases by State - Horizontal Bar")
plt.xlabel("Total Confirmed Cases"); plt.ylabel("State / UT")
plt.tight_layout(); plt.show()
```

```python
# Line Chart
plt.figure(figsize=(12,5))
plt.plot(latest['Name of State / UT'], latest['Total Confirmed cases'], marker='o', color='red')
plt.title("Total Confirmed Cases by State - Line Chart")
plt.xlabel("State / UT"); plt.ylabel("Total Confirmed Cases")
plt.xticks(rotation=90); plt.tight_layout(); plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(12,5))
plt.scatter(latest['Name of State / UT'], latest['Total Confirmed cases'], color='purple', s=80, alpha=0.7)
plt.title("State vs Total Confirmed Cases - Scatter Plot")
plt.xlabel("State / UT"); plt.ylabel("Total Confirmed Cases")
plt.xticks(rotation=90); plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout(); plt.show()
```

```python
#  Histogram
plt.figure(figsize=(10,5))
plt.hist(latest['Total Confirmed cases'], bins=15, color='skyblue', edgecolor='black')
plt.title("Distribution of Total Confirmed Cases - Histogram")
plt.xlabel("Total Confirmed Cases"); plt.ylabel("Number of States")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout(); plt.show()
```

```python
# Box Plot
plt.figure(figsize=(8,5))
plt.boxplot(latest['Total Confirmed cases'], vert=False, patch_artist=True,
            boxprops=dict(facecolor='orange', color='black'),
            medianprops=dict(color='blue', linewidth=2))
plt.title("Distribution of Total Confirmed Cases - Box Plot")
plt.xlabel("Total Confirmed Cases");
plt.tight_layout(); plt.show()
```

```python
# Pie Chart (Top 10 States)
top10 = latest.sort_values(by='Total Confirmed cases', ascending=False).head(10)
plt.figure(figsize=(8,8))
plt.pie(top10['Total Confirmed cases'], labels=top10['Name of State / UT'],
        autopct='%1.1f%%', startangle=140, colors=plt.cm.tab10.colors)
plt.title("Top 10 States by Total Confirmed Cases - Pie Chart")
plt.tight_layout(); plt.show()
```
