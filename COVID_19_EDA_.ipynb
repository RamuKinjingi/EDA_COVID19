```python
#Deleting two Columns
covid=data.drop(['Persons under Treatment including Home Treatment till 16.07.2020','Persons under Treatment including Home Treatment as on 17.07.2020'],axis=1)
```

```python
#Relationship Analysis
```

```python
#Line Graph on new cases and district
plt.figure(figsize=(14,6))
plt.plot(covid["District"], covid["New Cases on 17.07.2020"], marker='o', color='blue', linestyle='solid')
plt.title("Line Graph: District-wise New COVID-19 Cases")
plt.xlabel("District")
plt.ylabel("New Cases")
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
#Comparision Line Graph of new cases and death cases
plt.figure(figsize=(14,4))
plt.plot(covid["District"], covid["New Cases on 17.07.2020"], marker='o', color='orange', label='New Cases')
plt.plot(covid["District"], covid["Deaths on 17.07.2020"], marker='s', color='red', label='Death Cases')
plt.title("Comparison of New Cases and Discharged Cases by District")
plt.xlabel("District")
plt.ylabel("Number of Cases")
plt.xticks(rotation=90)
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Bar graph on district wise new cases
plt.figure(figsize=(14,6))
plt.bar(covid["District"], covid["New Cases on 17.07.2020"], color='blue')
plt.title("District-wise New COVID-19 Cases")
plt.xlabel("District")
plt.ylabel("New Cases")
plt.xticks(rotation=90)
plt.show()
```

```python
#Bar graph on district wise deaths
plt.figure(figsize=(14,6))
plt.bar(covid["District"], covid["Deaths on 17.07.2020"], color='orange')
plt.title("District-wise COVID-19 Deaths")
plt.xlabel("District")
plt.ylabel("Deaths")
plt.xticks(rotation=90)
plt.show()
```

```python
#Histogram
plt.figure(figsize=(8,5))
plt.hist(covid["New Cases on 17.07.2020"], bins=15, color='purple', edgecolor='black')
plt.title("Distribution of New Cases")
plt.xlabel("New Cases")
plt.ylabel("Number Of Districts")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(covid["New Cases on 17.07.2020"], covid["Cases Discharged on 17.07.2020"], color='green')
plt.title("New Cases vs Cases Discharged (17.07.2020)")
plt.xlabel("New Cases")
plt.ylabel("Cases Discharged")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.scatter(covid["New Cases on 17.07.2020"], covid["Deaths on 17.07.2020"], color='red')
plt.title("New Cases vs Deaths (17.07.2020)")
plt.xlabel("New Cases")
plt.ylabel("Deaths")
plt.grid(True)
plt.show()
```

```python
#OUTLIERS of the DATASET
plt.figure(figsize=(6,5))
plt.boxplot(covid["New Cases on 17.07.2020"])
plt.title("Boxplot of New COVID Cases")
plt.ylabel("New Cases")
plt.grid(True)
plt.show()
```
